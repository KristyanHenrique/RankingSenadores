```python
def cargos(todos,novos,codComissao):
    lista = pd.DataFrame()
    for i,dic in enumerate(novos):
        aux = pd.DataFrame.from_dict(dic,orient='index').T
        aux["codComissao"] = codComissao
        lista = pd.concat([aux,lista])
    lista = pd.concat([todos,lista])
    return lista
```
